# Working with dataset

```python
diamonds = pd.read_csv('diamonds_basket.csv')
diamonds.head()
```

```python
diamonds.info()
```

```python
diamonds[diamonds['items_id'].isna()]
```

```python
ids =  diamonds['items_id'].apply(lambda t: t.split(', '))
ids = list(ids)
ids
```

```python
import re
 
new_ids = []
for i in range(len(ids)):
    if len(ids[i]) > 1: 
        l = list()
        for y in ids[i]:
            if re.search(r'[A-Z]+_[0-9]+', y) is not None:
                l.append(re.search(r'[A-Z]+_[0-9]+', y).group())
        new_ids.append(l)
#    else: 
#        ids.remove(ids[i])
```

```python
print(len(ids), len(new_ids))
```

```python
# Python program to check if two
# to get unique values from list
# using numpy.unique
import numpy as np
  
# function to get unique values
  
  
def unique(list1):
    x = np.array(list1)
    print(np.unique(x))
        
print("the unique values from list is")
unique(new_ids)
```

```python
# Import permutations from the itertools module
from itertools import permutations

# Define the set of groceries
flattened = [i for t in new_ids for i in t]
setofitems = list(set(flattened))

len(setofitems)
```

```python
from matplotlib import pyplot as plt
## Creating distribution of Item Sold

Item_distr = pd.DataFrame(flattened, columns = ['items']).groupby(by = "items").size().reset_index(name='Frequency').sort_values(by = 'Frequency', ascending=False).head(30)

## Declaring variables

bars = Item_distr["items"]
height = Item_distr["Frequency"]
x_pos = np.arange(len(bars))

## Defining Figure Size

plt.figure(figsize=(16,9))

# Create bars
plt.bar(x_pos, height, color=(0.3, 0.4, 0.6, 0.6))

# Add title and axis names
plt.title("Top 10 Sold Items")
plt.xlabel("Item Name", )
plt.ylabel("Number of Quantity Sold")

# Create names on the x-axis
plt.xticks(x_pos, bars, rotation = 45)

# Show graph
plt.show()
```

```python
# Generate all possible rules from groceries list
rules = list(permutations(setofitems, 2))
```

```python
# Print the set of rules
#print(rules)

# Print the number of rules
print(len(rules))
```

```python
# Import the transaction encoder function from mlxtend
from mlxtend.preprocessing import TransactionEncoder

# Instantiate transaction encoder and identify unique items in transactions
encoder = TransactionEncoder().fit(new_ids)

# One-hot encode transactions
onehot = encoder.transform(new_ids)

# Convert one-hot encoded data to DataFrame
onehot = pd.DataFrame(onehot, columns = encoder.columns_)

# Print the one-hot encoded transaction dataset
print(onehot.head())
```

```python
# Compute the support
support = onehot.mean()

# Print the support
print(support.describe())
```

```python
# Compute the support
support = onehot.mean()
support = pd.DataFrame(support, columns=['support']).sort_values('support',ascending=False)

# Print the support
support.head()
```

```python
# Identifying frequent itemsets with Apriori
# Import apriori from mlxtend
from mlxtend.frequent_patterns import apriori

# Compute frequent itemsets using the Apriori algorithm
frequent_itemsets = apriori(onehot, 
                            min_support = 0.00555,
                            use_colnames = True)

# Print a preview of the frequent itemsets
print(frequent_itemsets.head())
```

```python
# Generating association rules

# Import the association rule function from mlxtend
from mlxtend.frequent_patterns import association_rules

# Compute all association rules for frequent_itemsets_1
rules = association_rules(frequent_itemsets, 
                            metric = "support", 
                         	min_threshold = 0.0015)

# Print association rules
print(rules)
```

```python
# Apply a 0.002 antecedent support threshold, 0.60 confidence threshold, and 2.50 lift threshold
filtered_rules = rules[(rules['antecedent support'] > 0.002) &
						(rules['consequent support'] > 0.01) &
						(rules['confidence'] > 0.6) &
						(rules['lift'] > 0.50)]

print(filtered_rules)
```

## Vizualizing

### Heatmaps

```python
# Visualizing itemset support
import seaborn as sns

# Compute frequent itemsets using a minimum support of 0.07
frequent_itemsets = apriori(onehot, min_support = 0.00555, 
                            use_colnames = True)

# Compute the association rules
rules = association_rules(frequent_itemsets, metric = 'support', 
                          min_threshold = 0.0)

# Replace frozen sets with strings
rules['antecedents'] = rules['antecedents'].apply(lambda a: ','.join(list(a)))
rules['consequents'] = rules['consequents'].apply(lambda a: ','.join(list(a)))

# Transform data to matrix format and generate heatmap
pivot = rules.pivot(index='consequents', columns='antecedents', values='support')
sns.heatmap(pivot)

# Format and display plot
plt.yticks(rotation=0)
plt.show()
```

```python
# Heatmaps with lift

# Import seaborn under its standard alias
import seaborn as sns

# Transform the DataFrame of rules into a matrix using the lift metric
pivot = rules.pivot(index = 'consequents', 
                   columns = 'antecedents', values= 'lift')

# Generate a heatmap with annotations on and the colorbar off
sns.heatmap(pivot, annot = True, cbar = False)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

### Scatterplots

```python
# Pruning with scatterplots

# Import seaborn under its standard alias
import seaborn as sns

# Apply the Apriori algorithm with a support value of 0.0075
frequent_itemsets = apriori(onehot, min_support = 0.00555, 
                            use_colnames = True)

# Generate association rules without performing additional pruning
rules = association_rules(frequent_itemsets, metric = 'support', 
                          min_threshold = 0.0)

# Generate scatterplot using support and confidence
sns.scatterplot(x = "support", y = "confidence", data = rules)
plt.show()

# Notice that the confidence-support border roughly forms a triangle. 
# This suggests that throwing out some low support rules would also mean that we would discard rules
# that are strong according to many common metrics.
```

```python
# Optimality of the support-confidence border

# Import seaborn under its standard alias
import seaborn as sns

# Apply the Apriori algorithm with a support value of 0.0075
frequent_itemsets = apriori(onehot, min_support = 0.00555, 
                         use_colnames = True)

# Generate association rules without performing additional pruning
rules = association_rules(frequent_itemsets, metric = "support", 
                          min_threshold = 0.0)

# Generate scatterplot using support and confidence
sns.scatterplot(x = "support", y = "confidence", 
                size = "lift", data = rules)
plt.show()
# If you look at the plot carefully, you'll notice that the highest values of lift 
# are always at the support-confidence border for any given value of supply or confidence.
```

### Parallel coordinates plot

```python
# Import the parallel coordinates plot submodule
from pandas.plotting import parallel_coordinates

# Compute the frequent itemsets
frequent_itemsets = apriori(onehot, min_support = 0.00555, 
                         use_colnames = True)

# Compute rules from the frequent itemsets with the confidence metric
rules = association_rules(frequent_itemsets)

# Convert rules into coordinates suitable for use in a parallel coordinates plot
rules['antecedent'] = rules['antecedents'].apply(lambda antecedent: list(antecedent)[0])
rules['consequent'] = rules['consequents'].apply(lambda consequent: list(consequent)[0])
rules['rule'] = rules.index
# Define coordinates and label
coords = rules[['antecedent','consequent','rule']]

# Generate parallel coordinates plot
parallel_coordinates(coords, 'rule')
plt.legend([])
plt.show()
```
